## Combine all the CSV files into one file.

```python
import csv
import pandas as pd
fout=open("out3.csv", "w")
header = ['Business', 'Address', 'City', 'Zip', 'State', 'Category', 'Latitude', 'Longitude']
fout.write(",".join(header))
fout.write("\n")
for num in range(1,20):      #Make sure you have max value in the range exceeds by 1 to the number of files. 
     for line in open("sh"+str(num)+".csv"):
        fout.write(line)  
fout.close()
```

## Sort the CSV files by Zip

```python
with open('out3.csv', 'r', newline='') as f_input:
    csv_input = csv.DictReader(f_input)
    data = sorted(csv_input, key=lambda row: (row['Zip']))

with open('output.csv', 'w', newline='') as f_output:    
    csv_output = csv.DictWriter(f_output, fieldnames=csv_input.fieldnames)
    csv_output.writeheader()
    csv_output.writerows(data)
```

```python
consolidated_data_to_load = "output.csv"
consolidated_data =  pd.read_csv(consolidated_data_to_load)
consolidated_data
```

```python
Business_group_by_zip = consolidated_data.groupby('Zip')
Number_of_Business_by_zip = pd.DataFrame({'Number of Business': Business_group_by_zip.size()}).reset_index()
Number_of_Business_by_zip
```

```python
Business_group_by_zip_business = consolidated_data.groupby(['Zip', 'Category'])
Number_of_Business_by_category_in_zip = pd.DataFrame({'Number of Business': Business_group_by_zip_business.size()}).reset_index()
Number_of_Business_by_category_in_zip
```
