## Set Up and Dependencies

```python
import json
import os
import csv
import glob
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# list of zip codes to be processed.
valid_zip = ['63101', '63102', '63103', '63104','63106', '63107', '63108', '63109', '63110', '63111','63112', '63113',
             '63115', '63116', '63118', '63139','63147']

# list for selecting/segragating businesses
save_list = ["Save", "Steak 'n Shake", "Schnucks", "Hospitals", "Urgent Care", "Medical Centers","Pediatricians"]
```

## Processing Starbucks JSON Data. 

```python
# Output File (CSV)
output_data_file = os.path.join("Resources", "Starbucks St.louis.csv")

# Load JSON
filepath = os.path.join("Resources", "starbucks.json")
with open(filepath) as jsonfile1:
    fastfood_json = json.load(jsonfile1)
    #business = len(businesses)
    #print("business", business)

# Read the JSON records and write the records to CSV file
with open(output_data_file, "w", encoding="utf8", newline="\n") as restaurant_file:
    writer = csv.writer(restaurant_file, delimiter=',')
    for business in fastfood_json.values():
         business_name1 = fastfood_json["businesses"]
         #print(business_name)
    
    for ffood in business_name1:
         fastfood = ffood["name"]
         ffood_categories = ffood['categories'][0]['title']     
         ffood_address = ffood['location']['address1']  
         ffood_city = ffood['location']['city']
         ffood_zip = ffood['location']['zip_code']  
         ffood_state = ffood['location']['state']  
         ffood_latitude = ffood['coordinates']['latitude']
         ffood_longitude = ffood['coordinates']['longitude']
         if ffood['location']['zip_code'] in valid_zip:    
             data1 = [fastfood, ffood_address, ffood_city, ffood_zip, ffood_state, ffood_categories, ffood_latitude, ffood_longitude]
             writer.writerow(data1)
             print(data1)
```
