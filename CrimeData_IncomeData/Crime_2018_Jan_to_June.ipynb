## Get the Crime category as a column. The first two numbers in the six digit crime code give the category. If there are only five number it is because the leading zero is cut off.

* 01  Homicide (Keep in mind that Excel will, by default, cut off the leading zero, so the number will look like 10000).
* 02  Rape (i.e. 20000  29999)
* 03  Robbery (i.e. 30000  39999)
* 04  Aggravated Assault
* 05  Burglary
* 06  Larceny
* 07  Vehicle Theft
* 08  Arson (08 and 09 are switched on the missouri list, but we will go with the st. louis list)
* 09  Other Assults
* 10  Forgery and Counterfeiting
* 11  Fraud
* 12  Embezzlement
* 13  Stolen Property
* 14  Vandalism
* 15  Weapons
* 16  Prositituion and Commericialized Vice
* 17  Sex Offenses (except rape, prositituion and commercialized vice)
* 18  Drug Abuse Violations
* 19  Gambling
* 20  Offenses Agains the Family and Children
* 21  Driving Under the Influence
* 22  Liquor Laws
* 23  Drunkenness 
* 24  Disorderly Conduct
* 25  Vagrancy
* 26  Other Offences

#### We will separate these into 
Violent Crimes (V):
01  09

Non-violent neighborhood Crimes (N):
13  18 and 20  25 

non-violent non-neighborhood Crimes (We will ignore these for our purposes):
10  12, 19 and 26


http://www.slmpd.org/Crime/CrimeDataFrequentlyAskedQuestions.pdf
https://ucr.mshp.dps.mo.gov/ucr/ucrhome.nsf/5cc4aefff80bfebe86256f3f005fd511/56ec43535549b5e7862573b7006e23b1/$FILE/State%20Agency%20UCR%20User%20Guide.pdf

```python
#Get the first 2 digits of the 6 digit crime code
crimes_total_df["Crime Category"]=crimes_total_df["Crime"].floordiv(10000)

#Make dataframe with neighborhoods for rows and 3 columns: V (violent), N (non-violent neighborhood), X (non neighborhood)
crimes_total_df["V/N/X"] =  np.where(crimes_total_df["Crime Category"]<10,"V",
                                   np.where(crimes_total_df["Crime Category"]<12, "X",#OR 
                                            np.where(crimes_total_df["Crime Category"]==19, "X", #OR 
                                                     np.where(crimes_total_df["Crime Category"]>25,"X","N"))))


#drop Neighborhood = 0 which is entries with incomplete addresses

#Used these to check number of crime data with zero neighborhood removed
crimes_total_df.shape
crimes_total_df.loc[crimes_total_df["Neighborhood"] == 0]

crimes_df = crimes_total_df[crimes_total_df.Neighborhood != 0]
crimes_df.shape
```

```python
#Make new dataframe of crime counts
neighborhood_crime_counts_df = pd.DataFrame(pd.pivot_table(crimes_df,values="Crime", index="Neighborhood", columns=["V/N/X"], aggfunc=len))
neighborhood_crime_counts_per_year_df=neighborhood_crime_counts_df
neighborhood_crime_counts_per_year_df["V"]=neighborhood_crime_counts_df["V"]/4.5
neighborhood_crime_counts_per_year_df["N"]=neighborhood_crime_counts_df["N"]/4.5
neighborhood_crime_counts_per_year_df["X"]=neighborhood_crime_counts_df["X"]/4.5
neighborhood_crime_counts_per_year_df
```

```python
#Get Neighborhood names
neighborhood_names_df = pd.read_csv("neighborhoods.csv")
neighborhood_names_df
#Merge dataframes
neighborhoods_df = pd.merge(neighborhood_names_df, neighborhood_crime_counts_per_year_df, on="Neighborhood")
neighborhoods_df.set_index("Neighborhood")
neighborhoods_df["crimes"]=neighborhoods_df["N"]+neighborhoods_df["V"]
neighborhoods_df = neighborhoods_df.sort_values(by=["crimes"])
neighborhoods_df
#neighborhoods_df.columns
```

```python
#Bar Chart for "V" and "N" crimes per neighborhood
N = len(neighborhoods_df.index)
violentCrimes = neighborhoods_df["V"]
nonviolentCrimes = neighborhoods_df["N"]
nonneighborhoodCrimes = neighborhoods_df["X"]

ind = np.arange(N)    # the x locations for the groups

#width = 2       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(5*ind, violentCrimes, width = 4, color='r')
p2 = plt.bar(5*ind, nonviolentCrimes, width = 4, bottom=violentCrimes,color = 'b')
p3 = plt.bar(5*ind, nonneighborhoodCrimes, width = 4, bottom=violentCrimes+nonviolentCrimes,color = 'g')

#plt.rcParams["figure.figsize"]=[20,10]

plt.xlim([-5,450])

plt.ylabel("Crime Count")
plt.title("St. Louis Crimes By Neighborhoods")
names = neighborhoods_df["Neighborhood Name"]
plt.xticks(5*ind, names,rotation=90)

#plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0], p3[0]), ("Violent Crimes", "Non-Voilent Crimes", "Non-Neighborhood Crimes"))

plt.plot(range(10))
plt.savefig('Results/crimeCount.png')
plt.show()
```

## Add zip codes (a neighborhood was labeled with the zip code where the majority of its land mass was) and 2010 population to the database of neighborhoods.

```python
zip_pop_df = pd.read_csv('STL2010CensusDemographicsByNeighborhood.csv')
zip_pop_df
```

```python
#Clean up zip_pop_df data
zip_pop_df = zip_pop_df.drop("Unnamed: 10", 1)
zip_pop_df = zip_pop_df.drop("Unnamed: 11", 1)
zip_pop_df = zip_pop_df.drop("Unnamed: 12", 1)
zip_pop_df = zip_pop_df.drop("Unnamed: 13", 1)
zip_pop_df = zip_pop_df.drop("Unnamed: 14", 1)
zip_pop_df = zip_pop_df.drop(79, 0)
zip_pop_df = zip_pop_df.drop(80, 0)

zip_pop_df["Neighborhood"]=zip_pop_df.Neighborhood.astype(int)
zip_pop_df["Population"]=zip_pop_df.Population.astype(int)
zip_pop_df
```

```python
#Add population and zip code data to crime data into one dataframe
neighborhood_crimes_df = pd.merge(neighborhoods_df, zip_pop_df, on="Neighborhood")
neighborhood_crimes_df=neighborhood_crimes_df.drop("Neighborhood Name_y",1)
neighborhood_crimes_df.rename({"Neighborhood Name_x": 'Neighborhood Name'}, axis = 1, inplace=True)
neighborhood_crimes_df.dtypes
neighborhood_crimes_df.columns

#Disregard neighborhoods that are in zip codes that are not entirely in the city. i.e. get rid of zip codes that are not numbers
stl_neighborhood_zip_crimes_df = neighborhood_crimes_df[neighborhood_crimes_df["Zip Code"]!="Not a city zip"]
stl_neighborhood_zip_crimes_df.dtypes
stl_neighborhood_zip_crimes_df.to_csv("Results/neighborhood_crime_data.csv")
stl_neighborhood_zip_crimes_df

```

```python
#Add columns which are neighborhood and violent crimes relative to neighborhood population.
stl_neighborhood_zip_crimes_df["Relative N Crimes"] = stl_neighborhood_zip_crimes_df["N"]/stl_neighborhood_zip_crimes_df["Population"]
stl_neighborhood_zip_crimes_df["Relative V Crimes"] = stl_neighborhood_zip_crimes_df["V"]/stl_neighborhood_zip_crimes_df["Population"]
stl_neighborhood_zip_crimes_df["Relative X Crimes"] = stl_neighborhood_zip_crimes_df["X"]/stl_neighborhood_zip_crimes_df["Population"]
stl_neighborhood_zip_crimes_df["Relative Total Crimes"]=stl_neighborhood_zip_crimes_df["Relative V Crimes"]+stl_neighborhood_zip_crimes_df["Relative N Crimes"]+stl_neighborhood_zip_crimes_df["Relative X Crimes"]

stl_neighborhood_zip_crimes_df.dtypes
```
